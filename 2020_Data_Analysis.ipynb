```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats
```

```python
OSMI_2020_df=pd.read_csv("Resources/OSMI 2020 Mental Health in Tech Survey Results .csv")
OSMI_2020_df.head()
```

```python
# Filter tech workers based on columns 'Is your employer primarily a tech company/organisation?'
tech_workers_2020 = OSMI_2020_df.loc[(OSMI_2020_df['Is your employer primarily a tech company/organization?'] == 1.0)]
tech_workers_2020
```

```python
# Treatment by country
country_df=tech_workers_2020[['Have you ever sought treatment for a mental health disorder from a mental health professional?','What country do you *live* in?']]
country_df.head()
```

```python
#Bar chart by lived in country
country_value_count = country_df['What country do you *live* in?'].value_counts()
country_value_count.plot(kind='bar',xlabel='Country',\
                        title='Count of Tech Workers Who Sought Treatment For Mental Health Issues Per Country 2020')
```

```python
# By country for diagnosis only
country_df=tech_workers_2020[['Have you ever been *diagnosed* with a mental health disorder?','What country do you *work* in?']]
```

```python
#Bar chart by worked in country
country_value_count = country_df['What country do you *work* in?'].value_counts()
country_value_count.plot(kind='bar',xlabel='Country',\
                        title='Count of Tech Workers with Diagnosed Mental Health Issues Per Country 2020')
```

```python
# Pie chart for Wellness Program.
wellness_count=tech_workers_2020['Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?'].value_counts()
wellness_count.plot(kind='pie',autopct='%1.1f%%',\
                          title='Count of Tech Workers who were involved in mental health discussions as part of a Wellness Campaign')
```
