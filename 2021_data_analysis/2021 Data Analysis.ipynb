```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats
import numpy as np
```

```python
df=pd.read_csv("OSMI 2021 Mental Health in Tech Survey Results .csv")
df.head()
```

```python
# Filter tech workers based on columns 'Is your employer primarily a tech company..'
tech_df_2021 = df.loc[df['Is your employer primarily a tech company/organization?'] == 1.0]
tech_df_2021
```

```python
#1 Bar charts by country using columns 'Have you ever sought treatment for..' & 'what coutry do you live..'as indicator
country_df_2021=tech_df_2021[['Have you ever sought treatment for a mental health disorder from a mental health professional?','What country do you *live* in?']]
country_df_2021.head()
```

```python
#Visualisation
country_count_2021 = country_df_2021['What country do you *live* in?'].value_counts()
country_count_2021.plot(kind='bar',xlabel='Country',\
                        title='Count of Tech Workers with Self Reported Mental Health Issues Per Country 2021')
plt.savefig('country_count_2021.png')
```

```python
#2. Pie chart by employer wellness programs 
wellness_count=tech_df_2021['Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?'].value_counts()
wellness_count.plot(kind='pie',autopct='%1.1f%%',\
                          title='Count of Tech Workers who were involved in mental health discussions as part of a Wellness Campaign')
```

```python
#4.Pie chart by family history of mental illness.

#1. current tech workers who have mental issue 
mental_issue_workers=tech_df_2021[tech_df_2021['Have you ever sought treatment for a mental health disorder from a mental health professional?']==1]

#2.family history distribution for ppl with mental issue 
family_history_distribution=mental_issue_workers['Do you have a family history of mental illness?'].value_counts()
family_history_distribution.plot(kind='pie',autopct='%1.1f%%',\
                          title='Family history investigation on tech Workers with reported mental issues')
pd.DataFrame(family_history_distribution)
```

```python
#5. Pie chart by workplace culture.
workplace_culture_count=tech_df_2021['Would you feel comfortable discussing a mental health issue \
with your direct supervisor(s)?'].value_counts()
workplace_culture_count.plot(kind='pie',autopct='%1.1f%%',\
                          title='Willingness to discuss mental health issue with employers')
plt.savefig('workplaceculture_2021.png')
```
